```python
import os
from threading import Lock

class DatabaseConnection:
    _instance = None
    _lock = Lock()

    def __init__(self, config_file_path):
        # Initialize with default values
        self._hostname = "localhost"
        self._username = "nursultan"
        self._password = "my_password_25"
        self._database_name = "data_nur"
        self.load_config(config_file_path)

    def load_config(self, config_file_path):
        if os.path.exists(config_file_path):
            with open(config_file_path, 'r') as f:
                config_data = f.readlines()
                for line in config_data:
                    parts = line.strip().split('=')
                    if len(parts) == 2:
                        key, value = parts
                        if key.strip() == 'hostname':
                            self._hostname = value.strip()
                        elif key.strip() == 'username':
                            self._username = value.strip()
                        elif key.strip() == 'password':
                            self._password = value.strip()
                        elif key.strip() == 'database_name':
                            self._database_name = value.strip()

    @staticmethod
    def get_instance(config_file_path):
        if DatabaseConnection._instance is None:
            with DatabaseConnection._lock:
                if DatabaseConnection._instance is None:
                    DatabaseConnection._instance = DatabaseConnection(config_file_path)
        return DatabaseConnection._instance

    def get_connection_info(self):
        return {
            'hostname': self._hostname,
            'username': self._username,
            'password': self._password,
            'database_name': self._database_name
        }

    def execute_query(self, query):
        # Placeholder for executing SQL queries
        print("Executing query:", query)

# Test program
if __name__ == "__main__":
    # Testing the Singleton pattern
    instance1 = DatabaseConnection.get_instance("config.ini")
    instance2 = DatabaseConnection.get_instance("config.ini")

    print("Are instance1 and instance2 the same object?", instance1 is instance2)

    # Testing connection info
    connection_info = instance1.get_connection_info()
    print("Connection info:")
    for key, value in connection_info.items():
        print(f"{key}: {value}")

    # Testing executing a query
    instance1.execute_query("SELECT * FROM table_name")
```
